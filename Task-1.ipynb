```python
#Loading Dataset 
import pandas as pd 
dataset = pd.read_csv("D:/Downloads/task-1/KaggleV2-May-2016.csv")
dataset
```

```python
dataset.info()
```

```python
#Handling Missing Values 
print(dataset.isnull().sum())  # View nulls

dataset.dropna(inplace=True) #drops null values 
```

```python
#Removing Dulicate Values 
dataset.drop_duplicates(inplace=True)
```

```python
#Standardizing 'Gender' and 'No-show' column
dataset['Gender'] = dataset['Gender'].str.upper().replace({'FEMALE': 'F', 'MALE': 'M'})

dataset['No-show'] = dataset['No-show'].str.upper().replace({'YES': 'Yes', 'NO': 'No'})
```

```python
dataset.head()
```

```python
# Convert appointment and scheduled dates into consistent type
dataset['ScheduledDay'] = pd.to_datetime(dataset['ScheduledDay']).dt.strftime('%d-%m-%Y')
dataset['AppointmentDay'] = pd.to_datetime(dataset['AppointmentDay']).dt.strftime('%d-%m-%Y')
```

```python
dataset.head()
```

```python
# Renaming columns to lowercase with underscores
dataset.columns = dataset.columns.str.lower().str.replace(' ', '_').str.replace('-', '_')
```

```python
dataset.head()
```

```python
# Convert dates to datetime type
dataset['scheduledday'] = pd.to_datetime(dataset['scheduledday'])
dataset['appointmentday'] = pd.to_datetime(dataset['appointmentday'])

# Format as dd-mm-yyyy and overwrite the original columns
dataset['scheduledday'] = dataset['scheduledday'].dt.strftime('%d-%m-%Y')
dataset['appointmentday'] = dataset['appointmentday'].dt.strftime('%d-%m-%Y')
# Convert age and remove invalid values
dataset = dataset[dataset['age'] >= 0]
dataset['age'] = dataset['age'].astype(int)

# Convert string categories to 'category' type
dataset['gender'] = dataset['gender'].str.upper().astype('category')
dataset['neighbourhood'] = dataset['neighbourhood'].str.strip().str.title().astype('category')
dataset['no_show'] = dataset['no_show'].str.strip().str.capitalize().astype('category')

# Convert binary columns to int
binary_cols = ['scholarship', 'hipertension', 'diabetes', 'alcoholism', 'handcap', 'sms_received']
dataset[binary_cols] = dataset[binary_cols].astype(int)
```
